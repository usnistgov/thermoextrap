```python
import thermoextrap.xtrapy.xpan_beta as xpan_beta
import xarray as xr
import numpy as np

def xr_test(a, b):
    xr.testing.assert_allclose(a, b.transpose(*a.dims))

x = np.random.rand(1000, 2, 3, 4)
u = np.random.rand(1000)

# I think this actually works
data0 = xpan_beta.factory_data(u, x, 3, val=['a','b','c'], central=True)

# or just convert to xarrays
xx = xr.DataArray(x, dims=['rec','a','b','c'])
uu = xr.DataArray(u, dims=['rec'])
data1 = xpan_beta.DataValuesCentral(uv=uu, xv=xx, order=3)

xr.testing.assert_allclose(data0.dxdu,data1.dxdu)

# yet another way
data2 = xpan_beta.DataCentralMomentsVals(uv=uu, xv=xx, order=3)
xr_test(data1.dxdu, data2.dxdu)

# and yet another
data3 = xpan_beta.DataCentralMomentsVals.from_vals(uv=u, xv=x, order=3, val=['a','b','c'])
xr_test(data1.dxdu, data3.dxdu)
```
